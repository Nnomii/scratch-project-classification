# Get features

```python
codeSmall = pd.read_csv('codeSmallv2.csv')

# Variable counts
sql_query = "select projectid, count(distinct parameter1) AS variableCount from codeSmall where \"block-type\" in ('readVariable', 'setVar:to:', 'showVariable:', 'changeVar:by:', 'hideVariable:')  group by projectid"
variableCounts = ps.sqldf(sql_query, locals())

# Say counts
sql_query = "select projectid, count(*) AS sayCount from codeSmall where \"block-type\" in ('say', 'say:duration:elapsed:from:')  group by projectid"
sayCounts = ps.sqldf(sql_query, locals())

# Total blocks
sql_query = "select projectid, count(*) AS totalCount from codeSmall group by projectid"
totalCounts = ps.sqldf(sql_query, locals())

# Merge dataframes on 'projectID'
merged_df = pd.merge(sayCounts, totalCounts, on='projectID')

# Calculate 'sayPercentage'
merged_df['sayPercentage'] = merged_df['sayCount'] / merged_df['totalCount'] * 100

# Say block percentage
sayPercentage = merged_df[['projectID', 'sayPercentage']]

# Cyclomatic complexity average
sql_query = "select projectID, \"script-rank\", count(*) + 1 as CC from codeSmall where \"block-type\" in ('doIfElse','doIf') group by projectID, \"script-rank\""
cc = ps.sqldf(sql_query, locals())
cc = cc.drop('script-rank', axis=1)
cc = cc.groupby('projectID', as_index=False)['CC'].mean()

# Loop count
sql_query = "select projectid, count(*) AS loopCount from codeSmall where \"block-type\" in ('doForever', 'doRepeat')  group by projectid"
loopCounts = ps.sqldf(sql_query, locals())

labels = pd.merge(labels, variableCounts, on='projectID', how='outer')
labels['variableCount'].fillna(0, inplace = True)
labels = pd.merge(labels, sayPercentage, on='projectID', how='outer')
labels['sayPercentage'].fillna(0, inplace = True)
labels = pd.merge(labels, cc, on='projectID', how='outer')
labels['CC'].fillna(1, inplace = True) # A program that has no branching statements like if-else or switch-case has a cyclomatic complexity of 1.
labels = pd.merge(labels, loopCounts, on='projectID', how='outer')
labels['loopCount'].fillna(0, inplace = True)

# print(labels)

counts = []
labelIDs = []
for e in labels.index:
    if labels['type'][e] == 'game':
        counts.append([labels['variableCount'][e], labels['sayPercentage'][e]])
        labelIDs.append(1)
    elif labels['type'][e] == 'animation':
        counts.append([labels['variableCount'][e], labels['sayPercentage'][e]])
        labelIDs.append(0)
    elif labels['type'][e] == 'interactiveart':
        counts.append([labels['variableCount'][e], labels['sayPercentage'][e]])
        labelIDs.append(0)
    else:
        continue
    
data = (np.array(counts), np.array(labelIDs))
print("Number of projects: ")
print(len(data[0]))
```


# Classifier comparison

A comparison of several classifiers in scikit-learn on synthetic datasets.
The point of this example is to illustrate the nature of decision boundaries
of different classifiers.
This should be taken with a grain of salt, as the intuition conveyed by
these examples does not necessarily carry over to real datasets.

Particularly in high-dimensional spaces, data can more easily be separated
linearly and the simplicity of classifiers such as naive Bayes and linear SVMs
might lead to better generalization than is achieved by other classifiers.

The plots show training points in solid colors and testing points
semi-transparent. The lower right shows the classification accuracy on the test
set.


```python
# Code source: Gal Varoquaux
#              Andreas Mller
# Modified for documentation by Jaques Grobler
# License: BSD 3 clause

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap

from sklearn.datasets import make_circles, make_classification, make_moons
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

names = [
#     "Linear SVM",
    "RBF SVM",
    "Decision Tree",
    "Random Forest",
]

classifiers = [
#     KNeighborsClassifier(3),
#     SVC(kernel="linear", C=0.025, random_state=42),
    SVC(gamma=2, C=1, random_state=42),
#     GaussianProcessClassifier(1.0 * RBF(1.0), random_state=42),
    DecisionTreeClassifier(max_depth=5, random_state=42),
    RandomForestClassifier(
        max_depth=5, n_estimators=10, max_features=1, random_state=42
    ),
#     MLPClassifier(alpha=1, max_iter=1000, random_state=42),
#     AdaBoostClassifier(random_state=42),
#     GaussianNB(),
#     QuadraticDiscriminantAnalysis(),
]

X, y = make_classification(
    n_features=2, n_redundant=0, n_informative=2, random_state=1, n_clusters_per_class=1
)
rng = np.random.RandomState(2)
X += 2 * rng.uniform(size=X.shape)
linearly_separable = (X, y)

datasets = [
#     make_moons(noise=0.3, random_state=0),
#     make_circles(noise=0.2, factor=0.5, random_state=1),
#     linearly_separable,
    data
]

figure = plt.figure(figsize=(27, 9))
i = 1
# iterate over datasets
for ds_cnt, ds in enumerate(datasets):
    # preprocess dataset, split into training and test part
    X, y = ds
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=42
    )

    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

    # just plot the dataset first
    cm = plt.cm.RdBu
    cm_bright = ListedColormap(["#FF0000", "#0000FF"])
    ax = plt.subplot(len(datasets), len(classifiers) + 1, i)
    if ds_cnt == 0:
        ax.set_title("Input data")
    # Plot the training points
    ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, edgecolors="k")
    # Plot the testing points
    ax.scatter(
        X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=0.6, edgecolors="k"
    )
    
    plt.ylabel('"Say" block percentage')
    plt.xlabel('Average cyclomatic complexity')
#     default_x_ticks = range(len(X_train[:, 0]))
#     plt.plot(default_x_ticks, y)
#     plt.xticks(default_x_ticks, x)
    
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)
    ax.set_xticks(())
    ax.set_yticks(())
    i += 1

    # iterate over classifiers
    for name, clf in zip(names, classifiers):
        ax = plt.subplot(len(datasets), len(classifiers) + 1, i)

        clf = make_pipeline(StandardScaler(), clf)
        clf.fit(X_train, y_train)
        score = clf.score(X_test, y_test)
        DecisionBoundaryDisplay.from_estimator(
            clf, X, cmap=cm, alpha=0.8, ax=ax, eps=0.5
        )

        # Plot the training points
        ax.scatter(
            X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, edgecolors="k"
        )
        # Plot the testing points
        ax.scatter(
            X_test[:, 0],
            X_test[:, 1],
            c=y_test,
            cmap=cm_bright,
            edgecolors="k",
            alpha=0.6,
        )
        
        plt.ylabel('"Say" block percentage')
        plt.xlabel('Average cyclomatic complexity')

        ax.set_xlim(x_min, x_max)
        ax.set_ylim(y_min, y_max)
        ax.set_xticks(())
        ax.set_yticks(())
        if ds_cnt == 0:
            ax.set_title(name)
        ax.text(
            x_max - 0.3,
            y_min + 0.3,
            ("%.3f" % score).lstrip("0"),
            size=20,
            horizontalalignment="right",
        )
        i += 1

plt.tight_layout()
plt.show()

# new input data (flowers) ?
# Iris dataset
```
